```python
import os
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt

from sentence_transformers import SentenceTransformer

from langchain_community.document_loaders import PyPDFLoader # 1.ë¡œë“œ
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter # 2.ì²­í¬
from langchain.embeddings import HuggingFaceEmbeddings # 3. ì„ë² ë”© ëª¨ë¸
from langchain.vectorstores import FAISS # 3. ë²¡í„° ì €ì¥
from langchain.retrievers import BM25Retriever, EnsembleRetriever # 4. ê²€ìƒ‰ ê¸°ë²•
from langchain_ollama import ChatOllama
from langchain.prompts import ChatPromptTemplate
from langchain.docstore.document import Document

from transformers import pipeline

from langchain import hub
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain

# ìƒëª…ë³´í—˜
# 1. ë¬¸ì„œ ë¡œë“œ ë° ì²­í¬ ë¶„í• 
loader = PyPDFLoader("í•œí™”ìƒëª… ê°„í¸ê°€ì… ì‹œê·¸ë‹ˆì²˜ ì•”ë³´í—˜(ê°±ì‹ í˜•) ë¬´ë°°ë‹¹_2055-001_002_ì•½ê´€_20220601_(2).pdf")
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=30
)
chunks = text_splitter.split_documents(documents)
chunks = chunks[11:]  # ë¶ˆí•„ìš”í•œ ì²­í¬ ì œì™¸ (ì˜ˆ: ëª©ì°¨)

import re

# ì²­í¬ ìƒì„± í›„ ì¤„ë°”ê¿ˆ ë° íŠ¹ìˆ˜ë¬¸ì ì²˜ë¦¬ (ë¬¸ì¥ ëì˜ ë§ˆì¹¨í‘œëŠ” ìœ ì§€)
chunks = [
    Document(
        page_content=re.sub(r'\s+', ' ', re.sub(r'[^\w\s.]', '', chunk.page_content)).strip(), 
        metadata=chunk.metadata
    ) for chunk in chunks
]


# ê° ì²­í¬ì— ID ë¶€ì—¬
for i, chunk in enumerate(chunks):
    chunk.metadata['doc_id'] = i

# 2. HuggingFace ì„ë² ë”© ë° FAISS ìƒì„±
embedding = HuggingFaceEmbeddings(model_name="BAAI/bge-m3")
vectorstore = FAISS.from_documents(documents=chunks, embedding=embedding)

# 3. í‚¤ì›Œë“œ ê²€ìƒ‰ì„ ìœ„í•œ BM25 ìƒì„±
bm25_retriever = BM25Retriever.from_documents(chunks)
bm25_retriever.k = 3

# 4. ì˜ë¯¸ ê²€ìƒ‰ + í‚¤ì›Œë“œ ê²€ìƒ‰ (ì•™ìƒë¸”)
top_k = 3
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": top_k})
ensemble_retrievers = [retriever, bm25_retriever]
ensemble_retriever = EnsembleRetriever(retrievers=ensemble_retrievers, weights=[0.7, 0.3])

# 5. LLM ì´ˆê¸°í™” ë° RAG ì²´ì¸ êµ¬ì„±
llm = ChatOllama(model="llama3.1", temperature=0.8, num_predict=300)

retrieval_qa_chat_prompt = ChatPromptTemplate.from_template("""
ë‹¤ìŒ ì»¨í…ìŠ¤íŠ¸ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì§ˆë¬¸ì— ë‹µë³€í•´ì£¼ì„¸ìš”. ì»¨í…ìŠ¤íŠ¸ì— ê´€ë ¨ ì •ë³´ê°€ ì—†ë‹¤ë©´,
"ì£¼ì–´ì§„ ì •ë³´ë¡œëŠ” ë‹µë³€í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤."ë¼ê³  ë§ì”€í•´ ì£¼ì„¸ìš”.

ì»¨í…ìŠ¤íŠ¸: {context}

ì§ˆë¬¸: {input}

ë‹µë³€:
""")

combine_docs_chain = create_stuff_documents_chain(llm, retrieval_qa_chat_prompt)
rag_chain = create_retrieval_chain(ensemble_retriever, combine_docs_chain)
```

```python
import gradio as gr

# ë³´í—˜ ì¢…ë¥˜ë³„ RAG ì²´ì¸ ì •ì˜ (ìƒì„±ëœ RAG ì²´ì¸ ì„¤ì •)
rag_chain_cancer = rag_chain  # ìƒëª…ë³´í—˜ RAG ì²´ì¸
rag_chain_life = rag_chain1  # ì†í•´ë³´í—˜ RAG ì²´ì¸
rag_chain_car = rag_chain2  # ìë™ì°¨ë³´í—˜ RAG ì²´ì¸

# ì„ íƒëœ RAG ì²´ì¸ì„ ì €ì¥í•  ë³€ìˆ˜
selected_chain = {"current": None}

def set_rag_chain(chain_type):
    """ì„ íƒëœ RAG ì²´ì¸ì„ ì„¤ì •í•˜ëŠ” í•¨ìˆ˜"""
    if chain_type == "ìƒëª…ë³´í—˜":
        selected_chain["current"] = rag_chain_cancer
    elif chain_type == "ì†í•´ë³´í—˜":
        selected_chain["current"] = rag_chain_life
    elif chain_type == "ìë™ì°¨ë³´í—˜":
        selected_chain["current"] = rag_chain_car

def submit_question(question, chat_history):
    try:
        print(f"ì§ˆë¬¸: {question}")  # ë””ë²„ê¹…

        # ì„ íƒëœ RAG ì²´ì¸ì´ ì—†ìœ¼ë©´ ì—ëŸ¬ ë°˜í™˜
        if selected_chain["current"] is None:
            return chat_history, "ë¨¼ì € ë³´í—˜ ì¢…ë¥˜ë¥¼ ì„ íƒí•˜ì„¸ìš”.", []

        # ì„ íƒëœ RAG ì²´ì¸ ì‹¤í–‰
        result = selected_chain["current"].invoke({"input": question})
        print(f"RAG ì²´ì¸ ê²°ê³¼: {result}")  # ë””ë²„ê¹…

        # ë‹µë³€ ë° ìƒìœ„ ë¬¸ì„œ ê°€ì ¸ì˜¤ê¸°
        answer = result.get("answer", "ë‹µë³€ì„ ìƒì„±í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        context_docs = result.get("context", [])[:3]  # ìƒìœ„ 3ê°œ ë¬¸ì„œë§Œ ê°€ì ¸ì˜¤ê¸°

        # ì±„íŒ… ê¸°ë¡ì— ì§ˆë¬¸ê³¼ ë‹µë³€ ì¶”ê°€
        chat_history.append((question, answer))

        # í•˜ì´ë¼ì´íŠ¸ ë°ì´í„°ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
        highlight_parts = []

        for doc in context_docs:
            doc_text = doc.page_content
            doc_source = doc.metadata.get("source", "ì•Œ ìˆ˜ ì—†ìŒ")
            doc_page = doc.metadata.get("page", "ì•Œ ìˆ˜ ì—†ìŒ")

            # ë¬¸ì„œë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ë¶„ë¦¬í•˜ê³ , ì „ì²˜ë¦¬ ì ìš©
            sentences = [s.strip() for s in doc_text.split('. ') if len(s.strip()) > 5]

            for sentence in sentences:
                print(f"ë¬¸ì¥: {sentence}")  # ë””ë²„ê¹…

                # í•˜ì´ë¼ì´íŠ¸ ì¡°ê±´ ìˆ˜ì •
                # ì¡°ê±´ 1: ë‹µë³€(answer)ì˜ ì£¼ìš” í‚¤ì›Œë“œê°€ ë¬¸ì¥ì— í¬í•¨ë  ê²½ìš°
                # ì¡°ê±´ 2: ë¬¸ì¥ ê¸¸ì´ê°€ ì ì ˆíˆ ê¸´ ê²½ìš°(ë„ˆë¬´ ì§§ì€ ê²½ìš° ì œì™¸)
                keywords = answer.split()[:5]  # ë‹µë³€ì˜ ì£¼ìš” í‚¤ì›Œë“œ ìµœëŒ€ 5ê°œ ì¶”ì¶œ
                if any(keyword in sentence for keyword in keywords) and len(sentence) > 20:
                    print(f"í•˜ì´ë¼ì´íŠ¸ ì¡°ê±´ ì¶©ì¡±: {sentence}")  # ë””ë²„ê¹…
                    highlight_parts.append((sentence, "highlight"))
                else:
                    highlight_parts.append((sentence, None))

            # ë¬¸ì„œì˜ ì¶œì²˜ ì •ë³´ëŠ” ë³„ë„ë¡œ ì¶”ê°€
            highlight_parts.append((f"\nì¶œì²˜: {doc_source}, í˜ì´ì§€: {doc_page}\n", None))

        # Gradioë¡œ ì˜¬ë°”ë¥´ê²Œ ë°ì´í„° ì „ë‹¬
        return chat_history, "", highlight_parts

    except Exception as e:
        print(f"ì˜¤ë¥˜ ë°œìƒ: {e}")  # ë””ë²„ê¹…
        import traceback
        traceback.print_exc()
        return chat_history, f"ì˜¤ë¥˜ ë°œìƒ: {e}", []

# ë‹¤ì‹œ ë³´ë‚´ê¸° ë²„íŠ¼ ë™ì‘ ì •ì˜
def retry_question(chat_history):
    """ì´ì „ ì§ˆë¬¸ì„ ë‹¤ì‹œ ì œì¶œí•˜ëŠ” í•¨ìˆ˜"""
    if not chat_history:
        return chat_history, "ë‹¤ì‹œ ì œì¶œí•  ì§ˆë¬¸ì´ ì—†ìŠµë‹ˆë‹¤.", []  # ëŒ€í™” ê¸°ë¡ì´ ì—†ìœ¼ë©´ ì—ëŸ¬ ì²˜ë¦¬

    last_question, _ = chat_history[-1]  # ë§ˆì§€ë§‰ ì§ˆë¬¸ ê°€ì ¸ì˜¤ê¸°
    return submit_question(last_question, chat_history[:-1])  # ì§ˆë¬¸ ì¬ì „ì†¡

# Gradio UI êµ¬ì„±
with gr.Blocks() as iface:
    # ì œëª© ë° ì„¤ëª…
    gr.Markdown("# ë³´í—˜ ë¬¸ì„œ ì±—ë´‡\në³´í—˜ ë¬¸ì„œì— ëŒ€í•´ ë¬¼ì–´ë³´ë©´ ë‹µí•˜ëŠ” ì±—ë´‡ì…ë‹ˆë‹¤.")

    # PDF ë²„íŠ¼ ë° ì±—ë´‡ UI
    with gr.Row():
        cancer_btn = gr.Button("ìƒëª…ë³´í—˜")
        life_btn = gr.Button("ì†í•´ë³´í—˜")
        car_btn = gr.Button("ìë™ì°¨ë³´í—˜")
        
    with gr.Row():  # ìš©ì–´ì‚¬ì „ ë²„íŠ¼
        glossary_btn = gr.Button("ìš©ì–´ì‚¬ì „ ğŸ“–")

    chatbot = gr.Chatbot()
    msg = gr.Textbox(placeholder="ë­ë“ ì§€ ë¬¼ì–´ë³´ì„¸ìš”.", label="ì±— ì…ë ¥")
    highlighted_text = gr.HighlightedText(label="ë¬¸ì„œ í•˜ì´ë¼ì´íŠ¸", show_legend=True)

    # ë²„íŠ¼ í–‰
    with gr.Row():
        submit_btn = gr.Button("ë³´ë‚´ê¸°")
        retry_btn = gr.Button("ë‹¤ì‹œë³´ë‚´ê¸° â†©")
        undo_btn = gr.Button("ì´ì „ ì±„íŒ… ì‚­ì œ âŒ")
        clear_btn = gr.Button("ì „ì²´ ì±„íŒ… ì‚­ì œ ğŸ’«")

    # ë²„íŠ¼ í´ë¦­ ì´ë²¤íŠ¸ ì„¤ì •
    def update_chatbot_with_message(chat_history, sender, message):
        chat_history.append((sender, message))  # ì§€ì •ëœ ë°œì‹ ìì™€ ë©”ì‹œì§€ ì¶”ê°€
        return chat_history

    # ë³´í—˜ ì„ íƒ ë²„íŠ¼ ì„¤ì •
    cancer_btn.click(
        lambda chat_history: (
            update_chatbot_with_message(chat_history, "ìƒëª…ë³´í—˜", "ìƒëª…ë³´í—˜ ê´€ë ¨ ì •ë³´ë¥¼ ë¬¼ì–´ë³´ì„¸ìš”."),
            set_rag_chain("ìƒëª…ë³´í—˜"),
        )[0],
        inputs=chatbot,
        outputs=chatbot,
    )
    life_btn.click(
        lambda chat_history: (
            update_chatbot_with_message(chat_history, "ì†í•´ë³´í—˜", "ì†í•´ë³´í—˜ ê´€ë ¨ ì •ë³´ë¥¼ ë¬¼ì–´ë³´ì„¸ìš”."),
            set_rag_chain("ì†í•´ë³´í—˜"),
        )[0],
        inputs=chatbot,
        outputs=chatbot,
    )
    car_btn.click(
        lambda chat_history: (
            update_chatbot_with_message(chat_history, "ìë™ì°¨ë³´í—˜", "ìë™ì°¨ë³´í—˜ ê´€ë ¨ ì •ë³´ë¥¼ ë¬¼ì–´ë³´ì„¸ìš”."),
            set_rag_chain("ìë™ì°¨ë³´í—˜"),
        )[0],
        inputs=chatbot,
        outputs=chatbot,
    )
    glossary_btn.click(
        lambda chat_history: update_chatbot_with_message(
            chat_history,
            "ìš©ì–´ì‚¬ì „",
            (
                "ìš©ì–´ì‚¬ì „ì— ëŒ€í•œ ì •ë³´ëŠ” ë‹¤ìŒ ë§í¬ë¥¼ ì°¸ì¡°í•˜ì„¸ìš”:\n"
                "- [ë³´í—˜ ìš©ì–´ì‚¬ì „ 1](https://www.hwgeneralins.com/lounge/dic/index.do)\n"
                "- [ë³´í—˜ ìš©ì–´ì‚¬ì „ 2](https://m.blog.naver.com/blogfsc/220149083103)"
            )
        ),
        inputs=chatbot,
        outputs=chatbot,
    )

    # ì§ˆë¬¸ ì œì¶œ ì²˜ë¦¬
    submit_btn.click(
        submit_question, 
        inputs=[msg, chatbot], 
        outputs=[chatbot, msg, highlighted_text]
    )
    
    # ë‹¤ì‹œ ë³´ë‚´ê¸° ë²„íŠ¼ ìˆ˜ì •
    retry_btn.click(
        retry_question,  # ì´ì „ ì§ˆë¬¸ ë‹¤ì‹œ ì œì¶œ
        inputs=[chatbot],  # ê¸°ì¡´ ì±„íŒ… ê¸°ë¡ì„ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©
        outputs=[chatbot, msg, highlighted_text]  # ê²°ê³¼ë¥¼ ì—…ë°ì´íŠ¸
    )

    # ì´ì „ ì±„íŒ… ì‚­ì œ ë²„íŠ¼
    undo_btn.click(
        lambda chat_history: (chat_history[:-1] if chat_history else [], "", []),  # í•˜ì´ë¼ì´íŠ¸ì™€ ë©”ì‹œì§€ë„ ì´ˆê¸°í™”
        inputs=[chatbot],
        outputs=[chatbot, msg, highlighted_text]
    )

    # ì „ì²´ ì±„íŒ… ì‚­ì œ ë²„íŠ¼
    clear_btn.click(
        lambda: ([], "", []),  # ì „ì²´ ì±„íŒ… ê¸°ë¡ ë° í•˜ì´ë¼ì´íŠ¸ ì´ˆê¸°í™”
        inputs=None,
        outputs=[chatbot, msg, highlighted_text]
    )

iface.launch(share=True, debug=True)
```
